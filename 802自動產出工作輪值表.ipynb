```python
!pip install workalendar
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    !pip install workalendarfrom openpyxl import Workbook
from openpyxl.styles import PatternFill, Border, Side, Alignment, Font
from openpyxl.utils import get_column_letter
from workalendar.asia import Taiwan
import calendar
from datetime import datetime

# === 年度設定 ===
year = 2025
minguo_year = year - 1911

# === 人員名單（依照你提供的圖） ===
staff_names = ["鄭重男", "劉建成", "李坤龍", "李庚儒", "陳采璇",
               "范玉玲", "林孟璇", "游曜銘", "許皇展", "吳松翰"]

# === 樣式設定 ===
border = Border(left=Side(style='thin'), right=Side(style='thin'),
                top=Side(style='thin'), bottom=Side(style='thin'))
center = Alignment(horizontal='center', vertical='center')
font_black = Font(color="000000", size=10)
font_white = Font(color="FFFFFF", size=10)
fill_white = PatternFill(fill_type=None)
fill_red = PatternFill(start_color="FF0000", end_color="FF0000", fill_type="solid")
fill_green = PatternFill(start_color="00B050", end_color="00B050", fill_type="solid")
fill_yellow = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# === 初始化台灣行事曆（國定假日） ===
cal = Taiwan()

# === 建立新的 Excel 工作簿 ===
wb = Workbook()
wb.remove(wb.active)

# === 為每個月份建立工作表 ===
for month in range(1, 13):
    ws = wb.create_sheet(title=f"{minguo_year}.{month:02d}資訊")
    ws["L1"] = f"國軍高雄總醫院(醫學資訊室){minguo_year}年{month:02d}月工作輪值表"
    ws.merge_cells("A1:B1")
    ws.merge_cells("L1:AA1")
    ws["A1"].alignment = center
    ws["L1"].alignment = center
    ws["A1"].font = Font(bold=True, size=14)
    ws["L1"].font = Font(bold=True, size=14)

    # === 日期與星期列 ===
    _, last_day = calendar.monthrange(year, month)
    # 加上項次、姓名欄位標籤
    ws["A2"] = "項次"
    ws["B2"] = "姓名/星期"
    ws["A2"].alignment = center
    ws["B2"].alignment = center
    ws["A2"].font = font_black
    ws["B2"].font = font_black
    ws["A2"].border = border
    ws["B2"].border = border
    for d in range(1, last_day + 1):
        col = get_column_letter(3 + d - 1)
        date = datetime(year, month, d)
        ws[f"{col}2"] = d
        ws[f"{col}3"] = "一二三四五六日"[date.weekday()]
        ws[f"{col}2"].alignment = center
        ws[f"{col}3"].alignment = center
        ws[f"{col}2"].font = font_black
        ws[f"{col}3"].font = font_black

    # === 人員與排班內容 ===
    for i, name in enumerate(staff_names):
        row = 4 + i
        ws[f"A{row}"] = i + 1
        ws[f"B{row}"] = name
        ws[f"A{row}"].alignment = center
        ws[f"B{row}"].alignment = center
        for d in range(1, last_day + 1):
            col = get_column_letter(3 + d - 1)
            weekday = datetime(year, month, d).weekday()
            cell = ws[f"{col}{row}"]

            if weekday == 6:
                cell.value = "休"
                cell.fill = fill_red
                cell.font = font_white
            elif weekday == 5:
                cell.value = "值"
                cell.fill = fill_green
                cell.font = font_white
            else:
                cell.value = "0800\n1700"
                cell.fill = fill_white
                cell.font = font_black
                cell.alignment = Alignment(horizontal='center', vertical='center', wrap_text=True)

            cell.alignment = Alignment(horizontal='center', vertical='center', wrap_text=True)

    # === 加入國定假日備註（A16起） ===
    holidays = cal.holidays(year)
    month_holidays = [(d, name) for d, name in holidays if d.month == month]
    for i, (d, name) in enumerate(month_holidays):
        cell = f"A{16 + i}"
        ws[cell] = f"【註】{d.month}月{d.day}日 {name}"
        ws[cell].fill = fill_yellow
        ws[cell].alignment = Alignment(wrap_text=True)
        ws[cell].font = Font(bold=True, size=10)

# === 儲存檔案 ===
wb.save("115年排班表_含節日備註.xlsx")
print(" 已成功產出 2026（民國115年）全年排班表，含節日註解！")

```

```python
!pip install python-docx

from docx import Document
from docx.shared import Cm # 導入 Cm 用於設定邊界
import calendar
import os
import shutil
from datetime import datetime, timedelta

# === 輸出資料夾 ===
output_dir = "/content/值班表_黑點依人員順序分配日期" # 如果在Colab上執行，使用Colab的路徑
# output_dir = r"C:\Users\kh802\Desktop\阿如交接\資訊室排班暨輪班表\值班表_黑點依人員順序分配日期2025624" # 如果在本地執行，取消註解這行並註解上面那行
os.makedirs(output_dir, exist_ok=True)

# 年度設定（114年＝2025年）
start_year = 2025
end_year = 2026 # 增加結束年份以處理跨年

# 輪值人員名單（會循環）- 這裡只放名字，用於輪值計算
personnel_names_only = [
    "劉建成", "李坤龍", "李庚儒", "陳采璇", "范玉玲",
    "林孟璇", "游曜銘", "許皇展", "吳松翰"
]

# 輪值人員名單（包含職稱，用於匹配聯絡資訊和值班表中的職稱+姓名欄位）
personnel_full_names = [
    "聘二-劉建成", "雇二-李坤龍", "聘三-李庚儒", "聘一-陳采璇", "聘二-范玉玲",
    "部工-林孟璇\n(請撥通資組726028)", "聘一-游曜銘", "聘三-許皇展", "聘三-吳松翰"
]

# 聯絡方式（範例資料）- 鍵仍然是包含職稱的完整名稱
phonebook = {
    "聘二-劉建成": ("0908-188-448", "0921-574-323", "0906-728-095", "728095"),
    "雇二-李坤龍": ("0972-369-081", "0972-369-081", "0906-728-093", "728093"),
    "聘三-李庚儒": ("0988-328-590", "0988-328-590", "0906-728-097", "728097"),
    "聘一-陳采璇": ("0938-122-236", "0938-122-236", "0906-728-094", "728094"),
    "聘二-范玉玲": ("0972-330-625", "0972-330-625", "0906-728-092", "728092"),
    "部工-林孟璇\n(請撥通資組726028)": ("0911-615-174", "0911-615-174", "0906-728-090", "728090"),
    "聘一-游曜銘": ("0973-358-096", "0973-358-096", "0972-056-139", "727760"),
    "聘三-許皇展": ("0903-050-656", "0903-050-656", "0972-056-105", "727757"),
    "聘三-吳松翰": ("0983-722-688", "0983-722-688", "0983-722-688", "727758")
}

# 建立從純名字到完整職稱姓名的映射，用於從輪值人員名字找到其完整的聯絡資訊鍵
name_to_full_name = {
    name.split('-')[-1].split('\n')[0].strip(): name # 提取名字部分作為鍵
    for name in personnel_full_names
}

# --- 計算起始輪值索引以符合 2026/1/3 由游曜銘開始 ---
target_person_on_jan_3_2026 = "游曜銘"
target_date_2026 = datetime(2026, 1, 3) # 目標日期 2026 年 1 月 3 日

try:
    # 找到游曜銘在純名字人員名單中的索引
    target_person_index = personnel_names_only.index(target_person_on_jan_3_2026)
except ValueError:
    print(f"錯誤：'{target_person_on_jan_3_2026}' 不在輪值人員名單中。請檢查 'personnel_names_only' 列表。")
    target_person_index = 0 # 如果找不到，預設從第一個人員開始

# 計算從 2025 年 1 月 1 日到 2026 年 1 月 3 日（包含 1 月 3 日）總共有多少個星期六
saturdays_until_target_date = 0
current_date_calc = datetime(start_year, 1, 1)

while current_date_calc <= target_date_2026:
    if current_date_calc.weekday() == 5: # 星期六是 5 (0=Mon, 1=Tue, ..., 5=Sat, 6=Sun)
        saturdays_until_target_date += 1
    current_date_calc += timedelta(days=1)

# `saturdays_until_target_date` 是從 2025 年 1 月 1 日算起，到 2026 年 1 月 3 日，總共經過了多少個星期六。
# 由於輪值索引是從 0 開始，所以當天是第 `(saturdays_until_target_date - 1)` 個輪值次數。
current_person_index_if_start_0 = (saturdays_until_target_date - 1) % len(personnel_names_only)

# 我們需要計算一個偏移量，使 `(實際次數 + 偏移量) % 總人數` 等於 `target_person_index`
# (target_person_index - current_person_index_if_start_0 + len(personnel_names_only)) % len(personnel_names_only)
rotation_offset = (target_person_index - current_person_index_if_start_0 + len(personnel_names_only)) % len(personnel_names_only)

# `global_rotation_start_index` 是整個長期輪值序列的起始索引
# 這表示 2025 年 1 月 1 日的第一個星期六 (即第 0 個輪值次數)，將會是 `personnel_names_only[rotation_offset]`
global_rotation_start_index = rotation_offset

print(f"計算出 {target_person_on_jan_3_2026} 值班 2026/1/3 所需的全局起始輪值索引：{global_rotation_start_index}")

# 輪值索引（跨月持續）
rotation_index = global_rotation_start_index

# === 每月 Word 產出 ===
for current_year in range(start_year, end_year + 1): # 處理 2025 和 2026 年
    start_month = 1
    end_month = 12
    if current_year == end_year:
        end_month = 1 # 2026 年只生成 1 月份的表

    for month in range(start_month, end_month + 1):
        doc = Document()
        minguo_year = current_year - 1911 # 更新民國年份

        # --- 調整頁面邊界和裝訂線 ---
        section = doc.sections[0]
        section.left_margin = Cm(1)
        section.right_margin = Cm(1)
        section.top_margin = Cm(1)
        section.bottom_margin = Cm(1)
        section.gutter = Cm(0)

        doc.add_heading(f"醫學資訊室 {minguo_year} 年 {month:02d} 月週六及假日值班表", level=1)
        doc.add_paragraph("※ 公務手機請隨身攜帶")

        # 當月所有週六
        cal = calendar.Calendar()
        # 修正語法錯誤： d for d in ...
        saturdays = [d for d in cal.itermonthdates(current_year, month) if d.weekday() == 5 and d.month == month]

        # 值班表
        doc.add_paragraph("1. 值班日期：")
        table = doc.add_table(rows=len(personnel_full_names) + 1, cols=len(saturdays) + 1)
        table.style = 'Table Grid'
        table.autofit = True
        table.allow_autofit = True

        table.cell(0, 0).text = "職稱-姓名"
        for col, day in enumerate(saturdays):
            table.cell(0, col + 1).text = str(day.day)

        for row, name_with_title in enumerate(personnel_full_names):
            table.cell(row + 1, 0).text = name_with_title
            for col in range(len(saturdays)):
                table.cell(row + 1, col + 1).text = ""

        # 指派輪值
        monthly_assignments_for_contact = [] # 儲存當月的值班資訊，用於聯絡方式表格

        # 臨時複製 rotation_index，避免影響下個月的起始計算
        # 不再需要臨時複製，因為 rotation_index 已經是全局累積的了
        # temp_rotation_index = rotation_index

        for col, day in enumerate(saturdays):
            person_name_only = personnel_names_only[rotation_index % len(personnel_names_only)]

            # 找到這個純名字對應的完整職稱姓名
            full_name_for_assignment = ""
            for p_full in personnel_full_names:
                if person_name_only in p_full: # 檢查純名字是否在完整名稱中
                    full_name_for_assignment = p_full
                    break

            if not full_name_for_assignment:
                 print(f"錯誤：找不到與純名字 '{person_name_only}' 匹配的完整職稱姓名。")
                 continue

            row_idx = personnel_full_names.index(full_name_for_assignment)
            table.cell(row_idx + 1, col + 1).text = "●"

            monthly_assignments_for_contact.append({
                'date': day,
                'person_name': person_name_only,
                'full_person_name': full_name_for_assignment
            })
            rotation_index += 1 # 全局輪值索引遞增

        # 聯絡方式
        doc.add_paragraph("\n2. 值班人員聯絡方式：")
        contact = doc.add_table(rows=1, cols=6)
        contact.style = 'Table Grid'
        contact.autofit = True
        contact.allow_autofit = True

        contact.rows[0].cells[0].text = "期間"
        contact.rows[0].cells[1].text = "輪值\n人員"
        contact.rows[0].cells[2].text = "GSM手機"
        contact.rows[0].cells[3].text = "家用電話"
        contact.rows[0].cells[4].text = "公務手機"
        contact.rows[0].cells[5].text = "簡碼"

        for assignment in monthly_assignments_for_contact:
            period_start = assignment['date']
            period_end = period_start + timedelta(days=6)

            period_str = f"{period_start.month:02d}{period_start.day:02d}—{period_end.month:02d}{period_end.day:02d}"

            person_name_to_display = assignment['person_name']
            full_person_name_for_lookup = assignment['full_person_name']

            gsm, home, gov, short = phonebook.get(full_person_name_for_lookup, ("-", "-", "-", "-"))

            row = contact.add_row().cells
            row[0].text = period_str
            row[1].text = person_name_to_display
            row[2].text = gsm
            row[3].text = home
            row[4].text = gov
            row[5].text = short

        # 備註與公告段落
        doc.add_paragraph(
        "\n上班期間：週一至週五 上午 8:00 至 下午 17:00，週六 上午 8:00 至 中午 12:00\n"
        "請注意：非上班期間，若值班同仁未接，則選撥當週值班同仁，組長：728086、代理副主任：728091\n"
        "本表交急診室掛號櫃台、衛整組訓練官、休假承辦人、護理部、注射室\n"
        "代理副主任：鄭重男\t\t\t日期：114.05.29\n"
        "擬：呈閱後公布\n"
        "醫學資訊室值班執行事曆"
        )
        filename = f"{minguo_year}年{month:02d}月值班表.docx"
        doc.save(os.path.join(output_dir, filename))

# === 壓縮為 zip ===
zip_path = shutil.make_archive(output_dir, 'zip', output_dir)
print(f"ZIP 已建立：{zip_path}")

# === Colab 下載區段 ===
try:
    from google.colab import files
    files.download(zip_path)
except ImportError:
    print("你不是在 Colab 上執行，請手動下載：", zip_path)
```
