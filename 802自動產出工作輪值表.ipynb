```python
!pip install python-docx

from docx import Document
from docx.shared import Cm # å°å…¥ Cm ç”¨æ–¼è¨­å®šé‚Šç•Œ
import calendar
import os
import shutil
from datetime import datetime, timedelta

# === è¼¸å‡ºè³‡æ–™å¤¾ ===
output_dir = "/content/å€¼ç­è¡¨_é»‘é»ä¾äººå“¡é †åºåˆ†é…æ—¥æœŸ" # å¦‚æœåœ¨Colabä¸ŠåŸ·è¡Œï¼Œä½¿ç”¨Colabçš„è·¯å¾‘
# output_dir = r"C:\Users\kh802\Desktop\é˜¿å¦‚äº¤æ¥\è³‡è¨Šå®¤æ’ç­æš¨è¼ªç­è¡¨\å€¼ç­è¡¨_é»‘é»ä¾äººå“¡é †åºåˆ†é…æ—¥æœŸ2025624" # å¦‚æœåœ¨æœ¬åœ°åŸ·è¡Œï¼Œå–æ¶ˆè¨»è§£é€™è¡Œä¸¦è¨»è§£ä¸Šé¢é‚£è¡Œ
os.makedirs(output_dir, exist_ok=True)

# å¹´åº¦è¨­å®šï¼ˆ114å¹´ï¼2025å¹´ï¼‰
start_year = 2025
end_year = 2026 # å¢åŠ çµæŸå¹´ä»½ä»¥è™•ç†è·¨å¹´

# è¼ªå€¼äººå“¡åå–®ï¼ˆæœƒå¾ªç’°ï¼‰- é€™è£¡åªæ”¾åå­—ï¼Œç”¨æ–¼è¼ªå€¼è¨ˆç®—
personnel_names_only = [
    "åŠ‰å»ºæˆ", "æå¤é¾", "æåºšå„’", "é™³é‡‡ç’‡", "èŒƒç‰ç²",
    "æ—å­Ÿç’‡", "æ¸¸æ›œéŠ˜", "è¨±çš‡å±•", "å³æ¾ç¿°"
]

# è¼ªå€¼äººå“¡åå–®ï¼ˆåŒ…å«è·ç¨±ï¼Œç”¨æ–¼åŒ¹é…è¯çµ¡è³‡è¨Šå’Œå€¼ç­è¡¨ä¸­çš„è·ç¨±+å§“åæ¬„ä½ï¼‰
personnel_full_names = [
    "è˜äºŒ-åŠ‰å»ºæˆ", "é›‡äºŒ-æå¤é¾", "è˜ä¸‰-æåºšå„’", "è˜ä¸€-é™³é‡‡ç’‡", "è˜äºŒ-èŒƒç‰ç²",
    "éƒ¨å·¥-æ—å­Ÿç’‡\n(è«‹æ’¥é€šè³‡çµ„726028)", "è˜ä¸€-æ¸¸æ›œéŠ˜", "è˜ä¸‰-è¨±çš‡å±•", "è˜ä¸‰-å³æ¾ç¿°"
]

# è¯çµ¡æ–¹å¼ï¼ˆç¯„ä¾‹è³‡æ–™ï¼‰- éµä»ç„¶æ˜¯åŒ…å«è·ç¨±çš„å®Œæ•´åç¨±
phonebook = {
    "è˜äºŒ-åŠ‰å»ºæˆ": ("0908-188-448", "0921-574-323", "0906-728-095", "728095"),
    "é›‡äºŒ-æå¤é¾": ("0972-369-081", "0972-369-081", "0906-728-093", "728093"),
    "è˜ä¸‰-æåºšå„’": ("0988-328-590", "0988-328-590", "0906-728-097", "728097"),
    "è˜ä¸€-é™³é‡‡ç’‡": ("0938-122-236", "0938-122-236", "0906-728-094", "728094"),
    "è˜äºŒ-èŒƒç‰ç²": ("0972-330-625", "0972-330-625", "0906-728-092", "728092"),
    "éƒ¨å·¥-æ—å­Ÿç’‡\n(è«‹æ’¥é€šè³‡çµ„726028)": ("0911-615-174", "0911-615-174", "0906-728-090", "728090"),
    "è˜ä¸€-æ¸¸æ›œéŠ˜": ("0973-358-096", "0973-358-096", "0972-056-139", "727760"),
    "è˜ä¸‰-è¨±çš‡å±•": ("0903-050-656", "0903-050-656", "0972-056-105", "727757"),
    "è˜ä¸‰-å³æ¾ç¿°": ("0983-722-688", "0983-722-688", "0983-722-688", "727758")
}

# å»ºç«‹å¾ç´”åå­—åˆ°å®Œæ•´è·ç¨±å§“åçš„æ˜ å°„ï¼Œç”¨æ–¼å¾è¼ªå€¼äººå“¡åå­—æ‰¾åˆ°å…¶å®Œæ•´çš„è¯çµ¡è³‡è¨Šéµ
name_to_full_name = {
    name.split('-')[-1].split('\n')[0].strip(): name # æå–åå­—éƒ¨åˆ†ä½œç‚ºéµ
    for name in personnel_full_names
}

# --- è¨ˆç®—èµ·å§‹è¼ªå€¼ç´¢å¼•ä»¥ç¬¦åˆ 2026/1/3 ç”±æ¸¸æ›œéŠ˜é–‹å§‹ ---
target_person_on_jan_3_2026 = "æ¸¸æ›œéŠ˜"
target_date_2026 = datetime(2026, 1, 3) # ç›®æ¨™æ—¥æœŸ 2026 å¹´ 1 æœˆ 3 æ—¥

try:
    # æ‰¾åˆ°æ¸¸æ›œéŠ˜åœ¨ç´”åå­—äººå“¡åå–®ä¸­çš„ç´¢å¼•
    target_person_index = personnel_names_only.index(target_person_on_jan_3_2026)
except ValueError:
    print(f"éŒ¯èª¤ï¼š'{target_person_on_jan_3_2026}' ä¸åœ¨è¼ªå€¼äººå“¡åå–®ä¸­ã€‚è«‹æª¢æŸ¥ 'personnel_names_only' åˆ—è¡¨ã€‚")
    target_person_index = 0 # å¦‚æœæ‰¾ä¸åˆ°ï¼Œé è¨­å¾ç¬¬ä¸€å€‹äººå“¡é–‹å§‹

# è¨ˆç®—å¾ 2025 å¹´ 1 æœˆ 1 æ—¥åˆ° 2026 å¹´ 1 æœˆ 3 æ—¥ï¼ˆåŒ…å« 1 æœˆ 3 æ—¥ï¼‰ç¸½å…±æœ‰å¤šå°‘å€‹æ˜ŸæœŸå…­
saturdays_until_target_date = 0
current_date_calc = datetime(start_year, 1, 1)

while current_date_calc <= target_date_2026:
    if current_date_calc.weekday() == 5: # æ˜ŸæœŸå…­æ˜¯ 5 (0=Mon, 1=Tue, ..., 5=Sat, 6=Sun)
        saturdays_until_target_date += 1
    current_date_calc += timedelta(days=1)

# `saturdays_until_target_date` æ˜¯å¾ 2025 å¹´ 1 æœˆ 1 æ—¥ç®—èµ·ï¼Œåˆ° 2026 å¹´ 1 æœˆ 3 æ—¥ï¼Œç¸½å…±ç¶“éäº†å¤šå°‘å€‹æ˜ŸæœŸå…­ã€‚
# ç”±æ–¼è¼ªå€¼ç´¢å¼•æ˜¯å¾ 0 é–‹å§‹ï¼Œæ‰€ä»¥ç•¶å¤©æ˜¯ç¬¬ `(saturdays_until_target_date - 1)` å€‹è¼ªå€¼æ¬¡æ•¸ã€‚
current_person_index_if_start_0 = (saturdays_until_target_date - 1) % len(personnel_names_only)

# æˆ‘å€‘éœ€è¦è¨ˆç®—ä¸€å€‹åç§»é‡ï¼Œä½¿ `(å¯¦éš›æ¬¡æ•¸ + åç§»é‡) % ç¸½äººæ•¸` ç­‰æ–¼ `target_person_index`
# (target_person_index - current_person_index_if_start_0 + len(personnel_names_only)) % len(personnel_names_only)
rotation_offset = (target_person_index - current_person_index_if_start_0 + len(personnel_names_only)) % len(personnel_names_only)

# `global_rotation_start_index` æ˜¯æ•´å€‹é•·æœŸè¼ªå€¼åºåˆ—çš„èµ·å§‹ç´¢å¼•
# é€™è¡¨ç¤º 2025 å¹´ 1 æœˆ 1 æ—¥çš„ç¬¬ä¸€å€‹æ˜ŸæœŸå…­ (å³ç¬¬ 0 å€‹è¼ªå€¼æ¬¡æ•¸)ï¼Œå°‡æœƒæ˜¯ `personnel_names_only[rotation_offset]`
global_rotation_start_index = rotation_offset

print(f"è¨ˆç®—å‡º {target_person_on_jan_3_2026} å€¼ç­ 2026/1/3 æ‰€éœ€çš„å…¨å±€èµ·å§‹è¼ªå€¼ç´¢å¼•ï¼š{global_rotation_start_index}")

# è¼ªå€¼ç´¢å¼•ï¼ˆè·¨æœˆæŒçºŒï¼‰
rotation_index = global_rotation_start_index

# === æ¯æœˆ Word ç”¢å‡º ===
for current_year in range(start_year, end_year + 1): # è™•ç† 2025 å’Œ 2026 å¹´
    start_month = 1
    end_month = 12
    if current_year == end_year:
        end_month = 1 # 2026 å¹´åªç”Ÿæˆ 1 æœˆä»½çš„è¡¨

    for month in range(start_month, end_month + 1):
        doc = Document()
        minguo_year = current_year - 1911 # æ›´æ–°æ°‘åœ‹å¹´ä»½

        # --- èª¿æ•´é é¢é‚Šç•Œå’Œè£è¨‚ç·š ---
        section = doc.sections[0]
        section.left_margin = Cm(1)
        section.right_margin = Cm(1)
        section.top_margin = Cm(1)
        section.bottom_margin = Cm(1)
        section.gutter = Cm(0)

        doc.add_heading(f"é†«å­¸è³‡è¨Šå®¤ {minguo_year} å¹´ {month:02d} æœˆé€±å…­åŠå‡æ—¥å€¼ç­è¡¨", level=1)
        doc.add_paragraph("â€» å…¬å‹™æ‰‹æ©Ÿè«‹éš¨èº«æ”œå¸¶")

        # ç•¶æœˆæ‰€æœ‰é€±å…­
        cal = calendar.Calendar()
        # ä¿®æ­£èªæ³•éŒ¯èª¤ï¼š d for d in ...
        saturdays = [d for d in cal.itermonthdates(current_year, month) if d.weekday() == 5 and d.month == month]

        # å€¼ç­è¡¨
        doc.add_paragraph("1. å€¼ç­æ—¥æœŸï¼š")
        table = doc.add_table(rows=len(personnel_full_names) + 1, cols=len(saturdays) + 1)
        table.style = 'Table Grid'
        table.autofit = True
        table.allow_autofit = True

        table.cell(0, 0).text = "è·ç¨±-å§“å"
        for col, day in enumerate(saturdays):
            table.cell(0, col + 1).text = str(day.day)

        for row, name_with_title in enumerate(personnel_full_names):
            table.cell(row + 1, 0).text = name_with_title
            for col in range(len(saturdays)):
                table.cell(row + 1, col + 1).text = ""

        # æŒ‡æ´¾è¼ªå€¼
        monthly_assignments_for_contact = [] # å„²å­˜ç•¶æœˆçš„å€¼ç­è³‡è¨Šï¼Œç”¨æ–¼è¯çµ¡æ–¹å¼è¡¨æ ¼

        # è‡¨æ™‚è¤‡è£½ rotation_indexï¼Œé¿å…å½±éŸ¿ä¸‹å€‹æœˆçš„èµ·å§‹è¨ˆç®—
        # ä¸å†éœ€è¦è‡¨æ™‚è¤‡è£½ï¼Œå› ç‚º rotation_index å·²ç¶“æ˜¯å…¨å±€ç´¯ç©çš„äº†
        # temp_rotation_index = rotation_index

        for col, day in enumerate(saturdays):
            person_name_only = personnel_names_only[rotation_index % len(personnel_names_only)]

            # æ‰¾åˆ°é€™å€‹ç´”åå­—å°æ‡‰çš„å®Œæ•´è·ç¨±å§“å
            full_name_for_assignment = ""
            for p_full in personnel_full_names:
                if person_name_only in p_full: # æª¢æŸ¥ç´”åå­—æ˜¯å¦åœ¨å®Œæ•´åç¨±ä¸­
                    full_name_for_assignment = p_full
                    break

            if not full_name_for_assignment:
                 print(f"éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°èˆ‡ç´”åå­— '{person_name_only}' åŒ¹é…çš„å®Œæ•´è·ç¨±å§“åã€‚")
                 continue

            row_idx = personnel_full_names.index(full_name_for_assignment)
            table.cell(row_idx + 1, col + 1).text = "â—"

            monthly_assignments_for_contact.append({
                'date': day,
                'person_name': person_name_only,
                'full_person_name': full_name_for_assignment
            })
            rotation_index += 1 # å…¨å±€è¼ªå€¼ç´¢å¼•éå¢

        # è¯çµ¡æ–¹å¼
        doc.add_paragraph("\n2. å€¼ç­äººå“¡è¯çµ¡æ–¹å¼ï¼š")
        contact = doc.add_table(rows=1, cols=6)
        contact.style = 'Table Grid'
        contact.autofit = True
        contact.allow_autofit = True

        contact.rows[0].cells[0].text = "æœŸé–“"
        contact.rows[0].cells[1].text = "è¼ªå€¼\näººå“¡"
        contact.rows[0].cells[2].text = "GSMæ‰‹æ©Ÿ"
        contact.rows[0].cells[3].text = "å®¶ç”¨é›»è©±"
        contact.rows[0].cells[4].text = "å…¬å‹™æ‰‹æ©Ÿ"
        contact.rows[0].cells[5].text = "ç°¡ç¢¼"

        for assignment in monthly_assignments_for_contact:
            period_start = assignment['date']
            period_end = period_start + timedelta(days=6)

            period_str = f"{period_start.month:02d}{period_start.day:02d}â€”{period_end.month:02d}{period_end.day:02d}"

            person_name_to_display = assignment['person_name']
            full_person_name_for_lookup = assignment['full_person_name']

            gsm, home, gov, short = phonebook.get(full_person_name_for_lookup, ("-", "-", "-", "-"))

            row = contact.add_row().cells
            row[0].text = period_str
            row[1].text = person_name_to_display
            row[2].text = gsm
            row[3].text = home
            row[4].text = gov
            row[5].text = short

        # å‚™è¨»èˆ‡å…¬å‘Šæ®µè½
        doc.add_paragraph(
        "\nä¸Šç­æœŸé–“ï¼šé€±ä¸€è‡³é€±äº” ä¸Šåˆ 8:00 è‡³ ä¸‹åˆ 17:00ï¼Œé€±å…­ ä¸Šåˆ 8:00 è‡³ ä¸­åˆ 12:00\n"
        "è«‹æ³¨æ„ï¼šéä¸Šç­æœŸé–“ï¼Œè‹¥å€¼ç­åŒä»æœªæ¥ï¼Œå‰‡é¸æ’¥ç•¶é€±å€¼ç­åŒä»ï¼Œçµ„é•·ï¼š728086ã€ä»£ç†å‰¯ä¸»ä»»ï¼š728091\n"
        "æœ¬è¡¨äº¤æ€¥è¨ºå®¤æ›è™Ÿæ«ƒå°ã€è¡›æ•´çµ„è¨“ç·´å®˜ã€ä¼‘å‡æ‰¿è¾¦äººã€è­·ç†éƒ¨ã€æ³¨å°„å®¤\n"
        "ä»£ç†å‰¯ä¸»ä»»ï¼šé„­é‡ç”·\t\t\tæ—¥æœŸï¼š114.05.29\n"
        "æ“¬ï¼šå‘ˆé–±å¾Œå…¬å¸ƒ\n"
        "é†«å­¸è³‡è¨Šå®¤å€¼ç­åŸ·è¡Œäº‹æ›†"
        )
        filename = f"{minguo_year}å¹´{month:02d}æœˆå€¼ç­è¡¨.docx"
        doc.save(os.path.join(output_dir, filename))

# === å£“ç¸®ç‚º zip ===
zip_path = shutil.make_archive(output_dir, 'zip', output_dir)
print(f"ZIP å·²å»ºç«‹ï¼š{zip_path}")

# === Colab ä¸‹è¼‰å€æ®µ ===
try:
    from google.colab import files
    files.download(zip_path)
except ImportError:
    print("ä½ ä¸æ˜¯åœ¨ Colab ä¸ŠåŸ·è¡Œï¼Œè«‹æ‰‹å‹•ä¸‹è¼‰ï¼š", zip_path)
```
